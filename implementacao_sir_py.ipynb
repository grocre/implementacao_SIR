```
import matplotlib.pyplot as plt

```

```
arrayS = [0]*25
arrayI = [0]*25
arrayR = [0]*25
arrayE = [0]*25

def sir(a = 0.001, b = 0.6, populacao = 1200, epocas = 25, infectados = 1):
  cont = 0
  while(cont < epocas):
    if(cont == 0): 
      s = (populacao - infectados) - (a*(populacao - infectados)*infectados)
      i = infectados + (a * s * infectados) - (b * infectados)
      r = 0 + i * b
      arrayS[cont] = s
      arrayI[cont] = i
      arrayR[cont] = r
      arrayE[cont] = cont

      
    else: 
      sAux = s;
      iAux = i;
      s = s - (a * s * i)
      i = i + (a * sAux * i) - b * i
      r = r + (b * iAux)
      arrayS[cont] = s
      arrayI[cont] = i
      arrayR[cont] = r
      arrayE[cont] = cont
    cont = cont + 1

  plt.plot(arrayE, arrayS, color="#17a589", label='SuscetÃ­veis')
  plt.plot(arrayE, arrayI, color="blue", label='Infectados')
  plt.plot(arrayE, arrayR, color="yellow", label='Recuperados')
  plt.ylim(0, 1200)
  
sir()

plt.xlabel("n")
plt.ylabel("S, I, R")
plt.plot(arrayE, arrayS)
plt.title('S, I, R x n   Infectados inicialmente = 1')
plt.legend()
```

```
plt.plot(arrayE, arrayS, label='SuscetÃ­veis')
plt.ylim(0,1300)
plt.xlabel('n')
plt.ylabel('SuscetÃ­veis')
plt.title('S x n')
plt.legend()
```

```
plt.plot(arrayE, arrayI, label='Infectados')
plt.ylim(0,1300)
plt.xlabel('n')
plt.ylabel('Infectados')
plt.title('I x n')
plt.legend()
```

```
plt.plot(arrayE, arrayR, label='Recuperados')
plt.ylim(0, 1300)
plt.xlabel('n')
plt.ylabel('Recuperados')
plt.title('R x n')
plt.legend()
```
